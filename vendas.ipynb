```
#importando as bibliotecas para calculos e planilhas e tambem o arquivo a ser trabalhado
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
importa_df = pd.read_csv('vgsales.csv')
importa_df.head()
```

```
#resumindo a planilha para 10000 colunas
vendas = importa_df.head(10000)
vendas.head()
```

```
#checando quantidade de linhas e colunas
vendas.shape
```

```
#removendo duplicatas
vendas = vendas.drop_duplicates()
vendas.shape
```

```
#listando os nomes das colunas
vendas.info()
```

```
#ajustando nomes de ingles para portugues Br
vendas_renomeado = vendas.rename(columns = {"Name": "Nome", "Platform": "Plataforma", "Year": "Ano", "Genre": "Genero", "NA_Sales": "Vendas_America_do_Norte", "EU_Sales": "Vendas_Europa", "Other_Sales": "Outras_Vendas", "Global_Sales":"Vendas_Global", "JP_Sales": "Vendas_Japão" })
vendas_renomeado.info()
```

```
vendas_renomeado.isnull().sum()
```

```
#preenchendo os valores N/A
vendas_renomeado['Ano'] = vendas_renomeado['Ano'].fillna('N/A')
vendas_renomeado['Publisher'] = vendas_renomeado['Publisher'].fillna('N/A')
```

```
#checando se está correto
vendas_renomeado.isnull().sum()
```

```
#com a base tratada verificar as estatisticas da planilha
vendas_renomeado.describe()
```

```
#Analise de vendas por região
vendas_regiao = vendas_renomeado.groupby(['Genero', 'Rank', 'Vendas_America_do_Norte', 'Vendas_Europa', 'Vendas_Japão', 'Outras_Vendas', 'Vendas_Global' ])['Rank'].count().reset_index(name='Qtd')
vendas_regiao
```

```
#analise grafica
fig = px.bar(vendas_regiao, x = 'Genero', y = 'Vendas_America_do_Norte',hover_data = ['Genero'], labels = {'Genero': 'Genero'}, color = 'Vendas_America_do_Norte')
fig.show()
```
